```python
%matplotlib inline

from collections import Counter, defaultdict
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from scipy.stats.stats import pearsonr
import cPickle as pkl
import datetime
import bisect
from pprint import pprint

import seaborn as sns
```

```python
rdata, udata, data, users, repos = pkl.load(open('data_structuring_150_pruned.pkl', 'rb'))
```

```python
all_times = []
for ukey, u in udata.iteritems():
    all_times += u['times']
```

```python
#all_times = sum([udata[u]['times'] for u in udata.keys()], [])
min_time = min(all_times)
max_time = max(all_times)
```

```python
def ts_to_buckets(ts_list, interval, end, grid):
    bucket_list = []
    for ts in ts_list:
        idx=bisect.bisect(grid,ts)
        bucket_list.append(idx)
    max_bucket = bisect.bisect(grid, end)
    return bucket_list, max_bucket
```

```python
interval = datetime.timedelta(days=3)
grid=[min_time+n*interval for n in range(300)] # this works for like many many years...
for ukey, u in udata.iteritems():
    u['time_buckets'] = ts_to_buckets(u['times'], interval, max_time, grid)[0]
for rkey, r in rdata.iteritems():
    r['time_buckets'] = ts_to_buckets(r['times'], interval, max_time, grid)[0]
for urkey, ur in data.iteritems():
    ur['time_buckets'], max_bucket = ts_to_buckets(ur['times'], 
                                                   interval, max_time, grid)
```
